```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Example data for visualizations
data = {
    'Fiscal Year': ['FY21', 'FY22', 'FY23'],
    'Net Sales': [76118, 80187, 82006],
    'Net Income': [14352, 14793, 14738],
    'EBIT Margin': [23.1, 22.4, 22.4],
    'Net Income Margin': [18.9, 18.4, 18.0],
    'Current Ratio': [0.70, 0.65, 0.63],
    'DSO': [22.3, 23.1, 24.0],
    'DIO': [58.0, 59.1, 59.5],
    'DPO': [133.1, 127.1, 122.9],
    'Net Trading Cycle': [-52.7, -44.9, -39.3],
    'Debt Ratio': [0.61, 0.60, 0.61],
    'Interest Coverage': [35.09, 40.99, 24.28],
    'ROA': [12, 13, 12],
    'ROE': [31, 32, 31],
    'COGS': [42760, 42157, 37108],
    'SG&A': [21112, 20217, 21024],
    'Total Assets': [119307, 117208, 120829],
    'Total Liabilities': [72653, 70354, 73764],
    'Shareholders\' Equity': [46654, 46854, 47065]
}

df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Fiscal Year'], df['Net Sales'], marker='o', label='Net Sales')
# plt.plot(df['Fiscal Year'], df['Net Income'], marker='o', label='Net Income')
plt.title('Revenue Trends')
plt.xlabel('Fiscal Year')
plt.ylabel('Amount ($ millions)')
plt.legend()
plt.grid(True)
plt.show()
```

The chart shows a steady increase in Net Sales from FY21 to FY23, indicating strong revenue growth over the period. The 2% growth from FY22 to FY23, although smaller than the 5% growth from FY21 to FY22, still reflects positive performance.

```python
df_margins = df[['Fiscal Year', 'EBIT Margin', 'Net Income Margin']].melt(id_vars='Fiscal Year', var_name='Margin Type', value_name='Margin (%)')

plt.figure(figsize=(10, 6))
sns.barplot(x='Fiscal Year', y='Margin (%)', hue='Margin Type', data=df_margins)
plt.title('Profitability Margins')
plt.xlabel('Fiscal Year')
plt.ylabel('Margin (%)')
plt.legend(title='Margin Type')
plt.show()
```

EBIT Margin: The EBIT margin has been consistent at 22.4% for both FY22 and FY23, down slightly from 23.1% in FY21. This stability indicates that P&G has managed its operational efficiency well, despite fluctuations in other areas.

Net Income Margin: The Net Income Margin decreased from 18.9% in FY21 to 18.0% in FY23. This slight decline may be attributed to higher expenses or lower efficiency, but the margin remains relatively high, showing effective control over net profitability.

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Fiscal Year'], df['Current Ratio'], marker='o', color='purple')
plt.title('Current Ratio Over Time')
plt.xlabel('Fiscal Year')
plt.ylabel('Current Ratio')
plt.grid(True)
plt.show()
```

Current Ratio: The Current Ratio decreased from 0.70 in FY21 to 0.63 in FY23. This decline suggests a reduction in P&Gs liquidity position, potentially indicating tighter working capital management or increased short-term liabilities relative to assets.

```python
df_efficiency = df[['Fiscal Year', 'DSO', 'DIO', 'DPO']].melt(id_vars='Fiscal Year', var_name='Efficiency Ratio', value_name='Days')

plt.figure(figsize=(12, 7))
sns.barplot(x='Fiscal Year', y='Days', hue='Efficiency Ratio', data=df_efficiency)
plt.title('Efficiency Ratios')
plt.xlabel('Fiscal Year')
plt.ylabel('Days')
plt.legend(title='Efficiency Ratio')
plt.show()
```

Days Sales Outstanding (DSO): DSO increased from 22.3 days in FY21 to 24.0 days in FY23, suggesting a slower collection of receivables. This could impact cash flow and working capital management.

Days Inventory Outstanding (DIO): DIO slightly increased over the years, indicating a small increase in the time P&G takes to sell its inventory. This could reflect changes in inventory management or market demand.

Days Payable Outstanding (DPO): DPO decreased from 133.1 days in FY21 to 122.9 days in FY23. A decrease in DPO suggests that P&G is paying its suppliers more quickly, which could be a sign of stronger supplier relationships or improved payment terms.

```python
plt.figure(figsize=(12, 7))
plt.plot(df['Fiscal Year'], df['Debt Ratio'], marker='o', label='Debt Ratio', color='blue')
# plt.plot(df['Fiscal Year'], df['Interest Coverage'], marker='o', label='Interest Coverage', color='red')
plt.title('Debt Ratio')
plt.xlabel('Fiscal Year')
plt.ylabel('Ratio')
plt.legend()
plt.grid(True)
plt.show()
```

Debt Ratio: The Debt Ratio remained relatively stable at 0.61 for FY21 and FY23, with a slight decrease to 0.60 in FY22. This stability indicates that P&G has managed its debt levels consistently in relation to its total assets.

```python
df_returns = df[['Fiscal Year', 'ROA', 'ROE']].melt(id_vars='Fiscal Year', var_name='Return Metric', value_name='Percentage')

plt.figure(figsize=(10, 6))
sns.barplot(x='Fiscal Year', y='Percentage', hue='Return Metric', data=df_returns)
plt.title('Return Metrics')
plt.xlabel('Fiscal Year')
plt.ylabel('Percentage (%)')
plt.legend(title='Return Metric')
plt.show()
```

Return on Assets (ROA): ROA decreased slightly from 13% in FY22 to 12% in FY23. This decline suggests that P&Gs efficiency in using its assets to generate profits has slightly weakened.

Return on Equity (ROE): ROE remained stable at 31% for FY21 and FY23, with a minor increase to 32% in FY22. This stability reflects consistent profitability relative to shareholders' equity.

```python
cost_data = {
    'Expense Type': ['COGS', 'SG&A'],
    'Amount ($ millions)': [df['COGS'].iloc[-1], df['SG&A'].iloc[-1]]
}

df_cost = pd.DataFrame(cost_data)

plt.figure(figsize=(8, 8))
plt.pie(df_cost['Amount ($ millions)'], labels=df_cost['Expense Type'], autopct='%1.1f%%', colors=['orange', 'lightblue'])
plt.title('Cost Management Breakdown')
plt.show()
```

Cost Distribution: The pie chart shows that COGS accounts for a significant portion of expenses compared to SG&A. This reflects the major cost drivers for P&G, with COGS being a substantial part of overall expenses. Effective cost management in these areas is crucial for maintaining profitability.

```python
df_balance = df[['Fiscal Year', 'Total Assets', 'Total Liabilities', 'Shareholders\' Equity']]

plt.figure(figsize=(12, 7))
df_balance.set_index('Fiscal Year').plot(kind='bar', stacked=True)
plt.title('Balance Sheet Composition')
plt.xlabel('Fiscal Year')
plt.ylabel('Amount ($ millions)')
plt.legend(title='Balance Sheet Components')
plt.show()
```

Asset and Liability Distribution: The stacked bar chart highlights the balance between Total Assets, Total Liabilities, and Shareholders' Equity over the years.

Total Assets: Increased from FY21 to FY23, indicating growth in the company's asset base.

Total Liabilities: Also increased, suggesting that P&G has taken on more debt or liabilities relative to its assets.
Shareholders Equity: Increased slightly, reflecting a stable increase in equity.

Overall Structure: The composition shows a well-balanced structure with a slight increase in equity supporting the growth in assets and liabilities.

P&G has shown steady revenue growth and stable profit margins, though liquidity and efficiency metrics suggest tighter cash flow and slower receivables collection. The company's debt ratio remains stable, but interest coverage has decreased, signaling higher financial pressure. Overall, P&G maintains a balanced balance sheet with effective cost management, supporting long-term financial stability.
